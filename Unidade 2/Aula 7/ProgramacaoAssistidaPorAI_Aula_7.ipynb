#### Passo 2 - Número Primo

```python
import math

def eh_primo(n):
    if n < 2: return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

# O Miniprojeto - Jogo da Forca
Um jogo da forca onde ao iniciar a execução, é gerada uma palavra aleatória e com isso é gerada também uma dica com relação à palavra gerada. Com a palavra gerada, é solicitado inputs do usuário pedindo uma letra para lançar tentativa de acerto a palavra. Havendo possibilidade de responder com a palavra inteira para confirmar acertos/erros. Não será possível acertar todas as letras para formar a palavra, haverá um limite para sobrar ao menos algumas letras faltando para então solicitar a palavra inteira ao usuário.

```python
import random

def jogar_forca():
    # Banco de palavras e dicas
    palavras_dicas = [
        ("PYTHON", "Linguagem de programação versátil"),
        ("ALGORITMO", "Sequência lógica de instruções"),
        ("COMPUTADOR", "Máquina para processamento de dados"),
        ("INTERNET", "Rede mundial de computadores"),
        ("VARIAVEL", "Espaço na memória para armazenar dados")
    ]

    # Escolha aleatória
    palavra, dica = random.choice(palavras_dicas)
    letras_descobertas = ["_" for _ in palavra]
    letras_tentadas = []

    # Limite para sobrar letras faltando (ex: não permite descobrir as últimas 2 letras via chute individual)
    LETRAS_RESTANTES_MINIMAS = 2

    print("=== BEM-VINDO AO JOGO DA FORCA ===")
    print(f"Dica: {dica}")

    while True:
        print(f"\nPalavra: {' '.join(letras_descobertas)}")
        print(f"Letras tentadas: {', '.join(letras_tentadas)}")

        # Verifica quantas letras faltam descobrir
        faltam = letras_descobertas.count("_")

        if faltam <= LETRAS_RESTANTES_MINIMAS:
            print("\nLimite de letras atingido! Agora você deve tentar adivinhar a palavra inteira.")
            chute_final = input("Qual é a palavra? ").strip().upper()
            if chute_final == palavra:
                print(f"Parabéns! Você acertou. A palavra era: {palavra}")
            else:
                print(f"Que pena! Você errou. A palavra correta era: {palavra}")
            break

        entrada = input("Digite uma letra ou a palavra inteira: ").strip().upper()

        if not entrada:
            continue

        # Se o usuário tentar a palavra inteira voluntariamente
        if len(entrada) > 1:
            if entrada == palavra:
                print(f"Incrível! Você acertou a palavra inteira: {palavra}")
            else:
                print(f"Erro fatal! A palavra não era '{entrada}'. Você perdeu. A correta era: {palavra}")
            break

        # Se o usuário tentar uma letra
        if entrada in letras_tentadas:
            print("Você já tentou essa letra!")
            continue

        letras_tentadas.append(entrada)

        if entrada in palavra:
            print(f"Boa! A letra '{entrada}' existe na palavra.")
            for i, letra in enumerate(palavra):
                if letra == entrada:
                    letras_descobertas[i] = entrada
        else:
            print(f"A letra '{entrada}' não existe na palavra.")

jogar_forca()
```

> *Gostaria que houvessem mais palavras, e que fosse possível validar uma letra ou palavra com a entrada que o usuário deu, para ser o quê deve ser independente da pontuação na letra. Isso vai facilidar um pouco o jogo. Mas também gostaria que essa mecânica fosse configurável no código através de uma variável.*

```python
import random
import unicodedata

def remover_acentos(texto):
    """Função auxiliar para remover acentos e caracteres especiais."""
    return "".join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')

def jogar_forca_melhorado():
    # CONFIGURAÇÃO: Se True, ignora acentos na validação (ex: 'Á' vira 'A')
    FACILITAR_ACENTOS = True

    # Banco de palavras ampliado
    palavras_dicas = [
        ("PYTHON", "Linguagem de programação versátil"),
        ("ALGORITMO", "Sequência lógica de instruções"),
        ("COMPUTADOR", "Máquina para processamento de dados"),
        ("INTERNET", "Rede mundial de computadores"),
        ("VARIAVEL", "Espaço na memória para armazenar dados"),
        ("DICIONARIO", "Estrutura de dados chave-valor"),
        ("BIBLIOTECA", "Conjunto de funções reutilizáveis"),
        ("REQUISICAO", "Pedido de dados a um servidor"),
        ("COMPILADOR", "Traduz código para linguagem de máquina"),
        ("ESTRUTURA", "Forma de organizar informações")
    ]

    palavra_original, dica = random.choice(palavras_dicas)

    # Define a versão que será usada para comparar os chutes
    if FACILITAR_ACENTOS:
        palavra_comparar = remover_acentos(palavra_original).upper()
    else:
        palavra_comparar = palavra_original.upper()

    letras_descobertas = ["_" for _ in palavra_original]
    letras_tentadas = []
    LETRAS_RESTANTES_MINIMAS = 2

    print("=== BEM-VINDO AO JOGO DA FORCA (VERSÃO 2.0) ===")
    print(f"Dica: {dica}")
    if FACILITAR_ACENTOS:
        print("Aviso: Acentos e pontuações serão ignorados nos seus palpites.")

    while True:
        print(f"\nPalavra: {' '.join(letras_descobertas)}")
        print(f"Letras tentadas: {', '.join(letras_tentadas)}")

        faltam = letras_descobertas.count("_")

        # Se faltarem poucas letras, exige a palavra inteira
        if faltam <= LETRAS_RESTANTES_MINIMAS:
            print("\nLimite de letras atingido! Tente adivinhar a palavra inteira.")
            chute_final = input("Qual é a palavra? ").strip().upper()

            chute_validacao = remover_acentos(chute_final) if FACILITAR_ACENTOS else chute_final

            if chute_validacao == palavra_comparar:
                print(f"Parabéns! Você acertou. A palavra era: {palavra_original}")
            else:
                print(f"Que pena! Você errou. A palavra correta era: {palavra_original}")
            break

        entrada = input("Digite uma letra ou a palavra inteira: ").strip().upper()
        if not entrada:
            continue

        # Normaliza o chute do usuário para comparação
        entrada_comparar = remover_acentos(entrada) if FACILITAR_ACENTOS else entrada

        # Verificação de tentativa de palavra inteira
        if len(entrada_comparar) > 1:
            if entrada_comparar == palavra_comparar:
                print(f"Incrível! Você acertou a palavra inteira: {palavra_original}")
            else:
                print(f"Erro! A palavra não era '{entrada}'. Fim de jogo. Era: {palavra_original}")
            break

        # Verificação de letra
        if entrada_comparar in letras_tentadas:
            print("Você já tentou essa letra!")
            continue

        letras_tentadas.append(entrada_comparar)

        if entrada_comparar in palavra_comparar:
            print(f"Boa! A letra '{entrada}' está na palavra.")
            for i, letra in enumerate(palavra_comparar):
                if letra == entrada_comparar:
                    letras_descobertas[i] = palavra_original[i]
        else:
            print(f"A letra '{entrada}' não existe na palavra.")

jogar_forca_melhorado()
```
